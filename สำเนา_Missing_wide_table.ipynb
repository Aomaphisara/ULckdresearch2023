Missing Data หรือ Missing Value
มี 3 ประเภท ได้แก่
1. MCAR หายไปแบบสุ่ม ข้อมูลที่หายไปไม่ขึ้นอยู่กับค่าใดค่าหนึ่งและเป็นอิสระต่อกัน
2. MAR หายไปแบบสุ่มในบางส่วน ข้อมูลที่หายไปขึ้นอยู่กับข้อมูลหรือตัวแปรอื่นๆ แต่ไม่ขึ้นอยู่กับค่าหรือตัวแปรของกลุ่มตัวเอง
3. MNAR หายไปแบบไม่สุ่ม ค่าของข้อมูลที่หายไปขึ้นอยู่กับค่าของข้อมูลในกลุ่มหรือตัวแปรเดียวกัน หรือรวมถึงตัวแปรอื่น ๆ ด้วย

```python
#ลบคอลัมน์ที่มีค่า Nan ทั้งหมดทิ้ง
df = df.dropna(how = 'all', axis = 1)
```

```python
#ลบแถวที่ซ้ำออก
df = df.drop_duplicates()
```

```python
df.shape
```

```python
missing_columns = df.columns[df.isnull().any()].tolist()
df_missing_columns = df[missing_columns]
```

```python
df_missing_columns.columns #แสดงเฉพาะคอลัมน์ที่มี missing เท่านั้น
```

```python
df_missing_columns.shape
```

```python
# คำนวณจำนวนค่าที่ขาดหายไปในแต่ละคอลัมน์
missing_values_count = df_missing_columns.isnull().sum()

# คำนวณเปอร์เซ็นต์ของค่าที่ขาดหายไปในแต่ละคอลัมน์
missing_values_percent = 100 * missing_values_count / len(df_missing_columns)

# สร้าง DataFrame ที่ประกอบด้วยจำนวนค่าที่ขาดหายไปและเปอร์เซ็นต์ของค่าที่ขาดหายไป
missing_data_info = pd.DataFrame({
    'Missing Values': missing_values_count,
    'Missing Values (%)': missing_values_percent
})

# แสดงผลลัพธ์ DataFrame ที่ถูกเรียงลำดับ
missing_data_info
```

```python
df_missing_columns[df_missing_columns.isnull().any(axis=1)] #ค้นหา missing ทั้งหมดทุกคอลัมน์
```

```python
df_missing_columns[df_missing_columns.isnull().any(axis=1)].index #หา index ของแถวที่มี missing data
```

```python
df.shape
```

```python
df_miss = df_missing_columns
```

```python
msno.bar(df_miss)
```

```python
msno.matrix(df)
```

```python
msno.matrix(df_miss)
```

```python
truncate = df_miss.isna().mean() > .025 # เลือกเฉพาะคอลัมน์ที่มีค่าหายไปมากกว่า 2.5%
pd.DataFrame({'sum': df_miss.loc[:, truncate].isna().sum(), 'mean': df_miss.loc[:, truncate].isna().mean()}).round(3).sort_values(by = 'mean', ascending = False) # เรียงลำดับตามค่าเฉลี่ยของค่าที่ขาดหายไปจากมากไปน้อย
```

```python
msno.heatmap(df)
```

```python
msno.dendrogram(df)
```

```python
msno.dendrogram(df_miss)
```

ดูจากค่าใน Heatmap
1. MCAR  (0) ไม่สัมพันธ์กับตัวแปรอื่น
2. MAR  (เข้าใกล้ -1) สัมพันธ์กับตัวแปรอื่นแต่ไม่สัมพันธ์กับตัวแปรของกลุ่มตัวเอง
3. MNAR  (เข้าใกล้ 1) สัมพันธ์กับตัวเเปรกลุ่มเดียวกัน
